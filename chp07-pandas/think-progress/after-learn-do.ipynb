```python
#缺失值用NAN代替
df7.fillna(np.nan, inplace=True)
#对异常值进行判断有多种方法,其中一种就是,当前值与均值的差值
#绝对值大于三倍标准差时,我们即将此值视为异常值
#axis=1与axis='columns'是等价的；跨越不同列即为:水平方向
#axis=0与axis='index'是等价的
for col in range(df7.shape[1]):
    df7.iloc[:, col][(df7.iloc[:, col] - df7.mean(axis = 1)).abs() > df7.std(axis=1) * 3] = np.nan
    
```

```python
#skew:偏度；kurt:峰度
df7.agg(["max", "min", "mean", "median", "sum", "std", "skew", "kurt"])
```

```python
#pandas中计算差分使用diff(self, periods=1, axis=0)
#periods:周期，默认值为1
#axis默认值为0或者index，表示在行之间做差分,如果值为
#1或columns,表示在列之前做差分
df7.diff(axis=1).agg(["max", "min", "mean", "median", "sum", "std", "skew", "kurt"])
```

```python
#获取每个用户用典数据最大值，最小值的索引月份
from collections import Counter
#df_max类型为pandas.core.series.Series
df_max = df7.max(axis=1)
df_min = df7.min(axis=1)
df_month_max = df_max.copy() #获取深拷贝
df_month_min = df_min.copy()
for row in range(df7.shape[0]):
    df_month_max.iloc[row] = Counter(df7.iloc[row, :][df7.iloc[row, :] 
                                                      == df_max.iloc[row]].index.month).most_common(1)[0][0]
    df_month_min.iloc[row] = Counter(df7.iloc[row, :][df7.iloc[row, :] 
                                                      == df_min.iloc[row]].index.month).most_common(1)[0][0] 
```

```python
#计算每个用户7月，8月用电量与3月，4月用电量和的比值，最大值的比值，最小值的比值，均值的比值
elec78 = df7.groupby(df7.columns.month, axis=1).sum().iloc[:, 6:8]
elec34 = df7.groupby(df7.columns.month, axis=1).sum().iloc[:, 2:4]
df_elec_sum_scale = elec78.sum(axis=1) / elec34.sum(axis=1)
df_elec_max_scale = elec78.max(axis=1) / elec34.max(axis=1)
df_elec_min_scale = elec78.min(axis=1) / elec34.min(axis=1)
df_elec_mean_scale = elec78.mean(axis=1) / elec34.mean(axis=1)
```

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

#生成正弦曲线点
nbSamples = 256
xRange = (-math.pi, math.pi) #X轴的取值区间[-π, +π]
x, y = [], []

for n in range(nbSamples):
    ratio = (n + 0.5) / nbSamples
    x.append(xRange[0] + (xRange[1] - xRange[0]) * ratio)
    y.append(math.sin(x[-1]))
    
plt.plot(x, y)
plt.show()
```

```python
#生成正弦曲线
import numpy as np
nbSamples = 256
x = np.linspace(-math.pi, math.pi, num=256)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#生成正弦曲线
import numpy as np
nbSamples = 256

x = np.linspace(-np.pi, np.pi, nbSamples)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='g', linewidth=4, linestyle='--')
plt.plot(x, y2, '*', markersize = 8, markerfacecolor='r', markeredgecolor='k')
```

```python
nbSamples = 128
x = np.linspace(-np.pi, np.pi, nbSamples)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color='g', linewidth=4, linestyle='--', label=r'$y=sin(x)$')
plt.plot(x, y2, '*', markersize = 8, markerfacecolor='r', markeredgecolor='k', label=r'$y=cos(x)$')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10,1)
y = x*2

for a,b in zip(x, y):
    plt.text(a, b, (a,b), ha='center', va='bottom', fontsize=10)

plt.plot(x, y, 'bo-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5,5, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)

#为在Matplotlib中显示中文, 设置特殊字体
#plt.rcParams['font.sns-serif'] = ['SimHei']
plt.title("双曲线")

plt.ylim(-1.2, 1.2)
plt.xlim(-6, 6)
plt.xticks(np.arange(-1.5*np.pi, 2*np.pi, 0.5 * np.pi))
plt.yticks([-1,0,1])
plt.xlabel('X')
plt.ylabel('Y')

plt.plot(x, y1, 'r-', label='$y_1=sin(x)$')
plt.plot(x, y2, 'b:', label='$y_2=cos(x)$')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0,5.0,0.2)

#第一种绘制子图的方法
fig = plt.figure() #创建一个画布
sub_fig1 = fig.add_subplot(211)
sub_fig1.grid(True)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

#第二种绘制子图的方法
plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
#生成一个画布,fig中有2*2分布均匀的子图
fig, axes_lst = plt.subplots(2,2)
x=np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
#在第1行第一列的子图中绘图
axes_lst[1,1].plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
left1, bottom1, with1, height1 = 0.1, 0.1, 0.8, 0.8
x=np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
#在画布上添加一个子图
axes_1 = fig.add_axes([left1, bottom1, with1, height1])
axes_1.scatter(x, y)
axes_1.set_xlabel('x')
axes_1.set_ylabel('y')
axes_1.set_title('title')
left2, bottom2, with2, height2 = 0.6, 0.6, 0.25, 0.25
#在画布上添加另外一个子图
axes_2 = fig.add_axes([2, bottom2, with2, height2])
axes_2.plot(x,y)
axes_2.set_title('title inside')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig = plt.figure()

axes_1 = fig.add_axes([0.1,0.1,0.5,0.5])
axes_2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes_3 = fig.add_axes([0.3,0.3,0.5,0.5])
axes_4 = fig.add_axes([0.4,0.4,0.5,0.5])
axes_4.plot(x, y)
plt.show()
```

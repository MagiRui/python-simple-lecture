```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('train.csv')
plt.rcParams[u'font.sans-serif']=['STHeiti']
plt.rcParams['axes.unicode_minus'] = False 
plt.figure(figsize=(10,5))
plt.subplot(121)
sns.countplot(x='Survived', data=train_df, hue='Survived')

plt.subplot(122)
_, _, autotexts = plt.pie([train_df.Survived.value_counts()[0],
                          train_df.Survived.value_counts()[1]],
                          labels=['幸存者', '非幸存者'],
                          autopct='%1.1f%%', explode=[0,0.05], 
                          shadow=True
                         )
#将饼图中的字体改成白色
for autotext in autotexts:
    autotext.set_color('white')

plt.title('幸存者数据统计')
plt.show()
```

```python
fig, [ax1, ax2, ax3] = plt.subplots(1, 3, figsize=(20,5))
sns.countplot(x='Sex', hue='Survived', data=train_df, ax=ax1)
sns.countplot(x='Pclass', hue='Survived', data=train_df, ax=ax2)
sns.countplot(x='Embarked', hue='Survived', data=train_df, ax=ax3)
ax1.set_title('Sex 特征分析')
ax2.set_title('Pclass 特征分析')
ax3.set_title('Embarked 特征分析')
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,5))
sns.countplot(x='SibSp', hue='Survived', data=train_df, ax=ax[0])
sns.countplot(x='Parch', hue='Survived', data=train_df, ax=ax[1])
ax[0].set_title('SibSp特征分析')
ax[1].set_title('Parch特征分析')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.kdeplot(train_df.loc[(train_df['Survived'] == 0), 'Age'],
           color='gray', linestyle='--', 
           shade=True, label='非幸存者')
sns.kdeplot(train_df.loc[(train_df['Survived'] == 1), 'Age'],
           color='gray', linestyle='--', 
           shade=True, label='幸存者')
plt.title('Age特征分布-Survivor V.S. Not Survivor', fontsize=15)
plt.xlabel("Age(年龄)", fontsize=15)
plt.ylabel('Frequency(频度)', fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.boxplot(x='Pclass', y="Age", data = train_df, ax = ax)
sns.swarmplot(x="Pclass", y = "Age", data = train_df, ax=ax, hue="Survived")
plt.show()
```

```python
df = pd.read_csv('us.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'], inplace=True)
df.head()
```

```python
df = pd.read_csv('us.csv', parse_dates=True, index_col=[0])
df.plot()
plt.show()
```

```python
style.use('ggplot') #换成ggplot的绘图风格
df.Illinois.plot() #绘制伊利诺伊(Illinois)周的流感趋势时序图
#df['Illinois'].plot()
plt.show()
```

```python
df.Illinois.plot()
df.Idaho.plot()#爱达荷周流感数据
plt.show()
```

```python
import matplotlib.pyplot as plt
style.use('default')
df.Illinois.plot(label='Illinois')
df.Idaho.plot(label='Idaho')
plt.legend(loc='best')
plt.show()
```

```python
temp_df = df[['Illinois','Idaho', 'Indiana']]
styles = ['bs-', 'ro-', 'y^-']
linewidths= [2,1,1]
labels = ['Illionis','Idaho', 'Indiana']
fig, ax = plt.subplots()
for col, style, lw, label in zip(temp_df.columns, styles, linewidths, labels):
    temp_df[col].plot(style=style, lw=lw, ax=ax, label=label)
plt.legend(loc='best')
plt.show()
```

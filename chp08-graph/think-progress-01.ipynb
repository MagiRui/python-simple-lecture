```python
#!/usr/bin/env python
# coding=utf-8
import numpy as np 
import matplotlib.pyplot as plt

plt.style.use('seaborn')
#
plt.rcParams[u'font.sans-serif']=['STHeiti']
plt.rcParams['axes.unicode_minus'] = False 

#
attr_labels = ['','','','','']
#
attr_size = len(attr_labels)
#
player = {
    '':[38,95,100,92,92],
    '':[72,94,91,96,96],
    '':[75,78,74,73,99],
    '':[83,65,65,79,81]
}

#4
plt.figure(figsize=(20,10))
ax1 = plt.subplot(221,projection='polar')
ax2 = plt.subplot(222,projection='polar')
ax3 = plt.subplot(223,projection='polar')
ax4 = plt.subplot(224,projection='polar')

#56
theta = np.linspace(0,2*np.pi,attr_size,endpoint=False)
theta = np.append(theta,theta[0])
for x in player:
    player[x] = np.append(player[x],player[x][0])

#
ax1.plot(theta,player[''],'c')
ax1.fill(theta,player[''],'c',alpha=0.3)#fill()
ax1.set_xticks(theta)
ax1.set_xticklabels(attr_labels,y=0.01,fontsize=15)
ax1.set_title('',color='c',fontsize=20)
ax1.set_yticks([])
ax1.grid(False)
for a,b in zip(theta,player['']):
    ax1.text(a,b,b,fontsize=10)


ax2.plot(theta,player[''],'m')
ax2.fill(theta,player[''],'m',alpha=0.3)
ax2.set_xticks(theta)
ax2.set_xticklabels(attr_labels,y=0.01,fontsize=15)
ax2.set_title('',color='m',fontsize=20)
ax2.set_yticks([])
ax2.grid(False)
for a,b in zip(theta,player['']):
     ax2.text(a,b,b,fontsize=10)

ax3.plot(theta,player[''],'r')
ax3.fill(theta,player[''],'r',alpha=0.3)
ax3.set_xticks(theta)
ax3.set_xticklabels(attr_labels,y=0.01,fontsize=15)
ax3.set_title('',color='r',fontsize=20)
ax3.set_yticks([])
ax3.grid(False)
for a,b in zip(theta,player['']):
    ax3.text(a,b,b,ha='center',fontsize=10)

ax4.plot(theta,player[''],'y')
ax4.fill(theta,player[''],'y',alpha=0.3)
ax4.set_xticks(theta)
ax4.set_xticklabels(attr_labels,y=0.01,fontsize=15)
ax4.set_title('',color='y',fontsize=20)
ax4.set_yticks([])
ax4.grid(False)
for a,b in zip(theta,player['']):
     ax4.text(a,b,b,ha='center',fontsize=10)

plt.tight_layout()
plt.show()
```
